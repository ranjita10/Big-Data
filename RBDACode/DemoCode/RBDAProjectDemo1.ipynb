```python
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
```

```python
arxiv_data = pd.read_csv("JasonFinal.csv", header=0, index_col=0)
library_data = pd.read_csv("LilyFinal.csv", header=0, index_col=0)
stack_data = pd.read_csv("RanjitaFinal.csv", header=0, index_col=0)
```

```python
def show_term(term, dataset=111):
    plt.figure(figsize=(16, 16), dpi=200)
    if dataset // 100 == 1 and term in arxiv_data.index:
        plt.plot(arxiv_data.loc[term], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term in library_data.index:
        plt.plot(library_data.loc[term], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term in stack_data.index:
        plt.plot(stack_data.loc[term], color="orange", label="stackoverflow.com")
    plt.title(term, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
```

```python
def show_both_term(term1, term2, dataset=111):
    plt.figure(figsize=(16, 8), dpi=200)
    plt.subplot(1,2,1)
    if dataset // 100 == 1 and term1 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term1], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term1 in library_data.index:
        plt.plot(library_data.loc[term1], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term1 in stack_data.index:
        plt.plot(stack_data.loc[term1], color="orange", label="stackoverflow.com")
    plt.title(term1, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
    axes = plt.gca()
    ymin, ymax = axes.get_ylim()
    plt.subplot(1,2,2)
    if dataset // 100 == 1 and term2 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term2], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term2 in library_data.index:
        plt.plot(library_data.loc[term2], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term2 in stack_data.index:
        plt.plot(stack_data.loc[term2], color="orange", label="stackoverflow.com")
    axes = plt.gca()
    axes.set_ylim(ymin, ymax)
    plt.title(term2, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
```

```python
def show_compare_term(term1, term2, dataset=111):
    plt.figure(figsize=(16, 16), dpi=200)
    if dataset // 100 == 1 and term1 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term1], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term1 in library_data.index:
        plt.plot(library_data.loc[term1], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term1 in stack_data.index:
        plt.plot(stack_data.loc[term1], color="orange", label="stackoverflow.com")
    plt.title(term1, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
    if dataset // 100 == 1 and term2 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term2], color="red",linestyle="--", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term2 in library_data.index:
        plt.plot(library_data.loc[term2], color="blue",linestyle="--", label="Libraries.io")
    if (dataset % 10) == 1 and term2 in stack_data.index:
        plt.plot(stack_data.loc[term2], color="orange",linestyle="--", label="stackoverflow.com")
    plt.title(term1 + " vs " + term2, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
```

```python
def show_four_term(term1, term2, term3, term4, dataset=111):
    plt.figure(figsize=(16, 16), dpi=200)
    plt.subplot(2,2,1)
    if dataset // 100 == 1 and term1 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term1], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term1 in library_data.index:
        plt.plot(library_data.loc[term1], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term1 in stack_data.index:
        plt.plot(stack_data.loc[term1], color="orange", label="stackoverflow.com")
    plt.title(term1, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
    axes = plt.gca()
    ymin, ymax = axes.get_ylim()
    plt.subplot(2,2,2)
    if dataset // 100 == 1 and term2 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term2], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term2 in library_data.index:
        plt.plot(library_data.loc[term2], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term2 in stack_data.index:
        plt.plot(stack_data.loc[term2], color="orange", label="stackoverflow.com")
    axes = plt.gca()
    axes.set_ylim(ymin, ymax)
    plt.title(term2, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
    plt.subplot(2,2,3)
    if dataset // 100 == 1 and term3 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term3], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term3 in library_data.index:
        plt.plot(library_data.loc[term3], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term3 in stack_data.index:
        plt.plot(stack_data.loc[term3], color="orange", label="stackoverflow.com")
    axes = plt.gca()
    axes.set_ylim(ymin, ymax)
    plt.title(term3, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
    plt.subplot(2,2,4)
    if dataset // 100 == 1 and term4 in arxiv_data.index:
        plt.plot(arxiv_data.loc[term4], color="red", label="arXiv.org")
    if (dataset % 100) // 10 == 1 and term4 in library_data.index:
        plt.plot(library_data.loc[term4], color="blue", label="Libraries.io")
    if (dataset % 10) == 1 and term4 in stack_data.index:
        plt.plot(stack_data.loc[term4], color="orange", label="stackoverflow.com")
    axes = plt.gca()
    axes.set_ylim(ymin, ymax)
    plt.title(term4, fontsize='xx-large')
    plt.legend(loc='upper right', fontsize='xx-large')
```
